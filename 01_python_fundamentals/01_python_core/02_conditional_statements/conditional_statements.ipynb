# Operators in python
#### There are 7 types of operators in python
1. Arithmetic Operators
2. Comparison Operators
3. Logical Operators
4. Bitwise Operators
5. Assignment Operators
6. Identity Operators
7. Membership Operators

# 1. Arithmatic Operators

```
# addition
2 + 3
```

```
# subtraction
7 - 2
```

```
# multiplication
8 * 2
```

```
# division
# returns float
16 / 2
```

```
# floor division
# returns int
16 // 3
```

```
# modulus
# returns remainder
16 % 3
```

```
# power
# returns 2 power 3
2**3
```

# 2. Comparison Operators

```
# greater than operator
7 > 3
```

```
# greater than or equal to
10 >= 5
```

```
# less than operator
3 < 7
```

```
# less than or equal to
5 <= 10
```

```
# equal to opeartor
10 == 10
```

```
# not equal to operator
10 != 20
```

# 3. Logical Operators

```
# OR operator
# or
True or False
```

```
# AND operator
# and
True and False
```

```
# NOT operator
# not
not True
```

# 4. Bitwise Operators

```
# Bitwise NOT operator
~10
```

```
# Bitwise OR
# performs bit wise OR operation
# 10 -> 1 0 1 0
# 04 -> 0 1 0 0
# Result
# 14 -> 1 1 1 0
10 | 4
```

```
# Bitwise AND
# performs bit wise AND operation
# 10 -> 1 0 1 0
# 04 -> 0 1 0 0
# Result
# 00 -> 0 0 0 0
10 & 4
```

```
# Bitwise XOR
# performs bit wise XOR operation
# 10 -> 1 0 1 0
# 04 -> 0 1 0 0
# Result
# 00 -> 0 0 0 0
10 ^ 4
```

```
# Left Shift
# format a << n
# add n number of 0 bits at the end of the a
# 10 << 2 i.e., 1 0 1 0 0 0 which results to 40
# 10 * 2 power n
10 << 2
```

```
# Right Shift
# format a >> n
# add n number of 0 bits at the end of the a
# 10 >> 2 i.e., 0 0 1 0 which results to 2
# 10 / 2 power n
10 >> 2
```

# 5. Assignment Operators

```
# a=n
# assign n to variable a
a = 40
a
```

```
# a += 10
# Add and Assign
# i.e., a=a+10
a += 10
a
```

```
# a -= 10
# Subtract and Assign
# i.e., a=a-10
a -= 10
a
```

# 6. Identity Operators

```
# is
# is not
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
# returns True because z is the same object as x

print(x is y)
# returns False because x is not the same object as y, even if they have the same content

print(x == y)
# to demonstrate the difference betweeen "is" and "==": this comparison returns True because x is equal to y
```

# 7. Membership Operators

```
# in
a = "hello world"
"hello" in a
```

```
# not in
"hello" not in a
```

# Conditional statements

```
a = 50
if a < 50:
    print("a is less than 50")
elif a > 50:
    print("a is greater than 50")
else:
    print("a is 50")
```

# Loops

- for loop
- for else loop
- while loop
- while else

##### keywords in loops 
- pass
- continue
- break

```
for i in range(1, 10):
    print(i, end=" ")
```

```
for i in range(1, 10):
    if i == 5:
        pass
    print(i, end=" ")
```

```
for i in range(1, 10):
    if i == 5:
        break
    print(i, end=" ")
```

```
for i in range(1, 10):
    if i == 5:
        print()
        continue
    print(i, end=" ")
```

```
i = 5
while i > 0:
    print(i)
    i -= 1
```
