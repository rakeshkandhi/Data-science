## Slicing operation 
#### The slicing operation is same for strings, lists and tuples
- variable[start, stop, step] 
- <strong>Start</strong>: Start index. Default from start
- <strong>Stop</strong>: Stop index. Default till end
- <strong>Step</strong>: Step specifies to increase the count for each cycle. Default 1

```
example = [12, 23, 34, 232, 45, 65, 767, 76, 45, 68]
```

```
# to get the elements from starting to index 3
# example[0:4]
example[:4]
```

```
# to get the elements from index 5 to end
# example[5:-1]
# example[5:]
example[5 : len(example)]
```

```
# to get the elements from index 2 to index 8
example[2:8]
```

```
# to get the elements from index 2 to index 8 with step 2
example[2:8:2]
```

```
# to get he element last 5 elements
example[-6:-1]
```

```
# reverse using the slicing
example[::-1]
```

## Map method

```
def square(n):
    return n * n
```

```
import numpy as np

lst = np.arange(1, 20, dtype="int32")
lst
```

```
for i in range(len(lst)):
    print(square(lst[i]), end=", ")
```

```
len(list(map(square, lst)))
```

```
print(list(map(lambda a: a * a, lst)), sep=" ")
```
